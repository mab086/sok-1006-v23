## Kapittel 3 - Bilisters tilpasning til bompenger

```python
def budsjett(M, p1, p2, x1):
    return M/p2 - p1*x1/p2

x = np.linspace(0.001, 1000, 1000)

def create_ax():
    fig, ax = plt.subplots()
    ax.set_ylabel('$x_2$', loc='top', rotation=0)
    ax.set_xlabel('$x_1$', loc='right')
    ax.set_xlim(0, 1000)
    ax.set_ylim(0, 2000)
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    return fig, ax

fig, ax = create_ax()

ax.plot(x, budsjett(1766, 3.61, 1, x), label='Konsumentens budsjett før bompenger', color='blue')
ax.plot(x, budsjett(1766, 3.03, 1, x), label='Konsumentens budsjett etter bompenger', color='red')

ax.fill_between(x, budsjett(1766, 3.61, 1, x), budsjett(1766, 3.03, 1, x), color="pink", alpha=0.3, label='Redusert mulighetsområde')

ax.set_title('Budsjettbetingelse')
ax.legend(loc='best')

```

I figuren har jeg presentert en illustrasjon av budsjettbetingelsen ved hjelp av de gitte estimatene. Budsjettlinjene representerer de økonomiske begrensningene en konsument står overfor når de må velge mellom antall kilometer kjørt og andre goder. Skjæringspunktene mellom budsjettlinjene og x1-aksen indikerer den maksimale mengden konsumenten har råd til å bruke på godet, forutsatt at hele budsjettet blir benyttet. Det samme gjelder for skjæringspunktene med x2-aksen. For å finne skjæringspunktet med x1-aksen, setter vi x2 = 0.

Konsumentens handlingsrom består av alle mulige kombinasjoner av x1 og x2 som ligger innenfor eller på budsjettlinjen. Punkter utenfor dette handlingsrommet er utilgjengelige for konsumenten. I figur 3 kan vi observere endringer i budsjettlinjen før og etter innføringen av bompenger for en bilist. Vi legger merke til en forskyvning av budsjettlinjen, noe som medfører en reduksjon i konsumentens handlingsrom som en konsekvens av økte kostnader ved bilkjøring. Denne reduksjonen er visualisert som det rosa området. Budsjettlinjen endrer helning og blir brattere. Skjæringspunktet med x1-aksen flyttes mot venstre, og brøken reduseres. Økningen i kostnadene for godet 1 har en negativ innvirkning på konsumentens kjøpekraft. Dette temaet vil bli ytterligere utforsket senere i oppgaven.

```python
# Symboler
x_1, x_2, a, b, U_0 = sp.symbols("x_1 x_2 a b U_0", real=True, positive=True)
U = x_1**a * x_2**b

# Løsning for x_2 ved gitt nyttenivå
x_2_ind_sol = sp.solve(U - U_0, x_2)[0]

# Lambda-funksjon for å plotte
indiff_x_2 = sp.lambdify((U_0, a, b, x_1), x_2_ind_sol)

# Nye symboler
p_1, p_2, m = sp.symbols("p_1 p_2 m", real=True, positive=True)

# Budsjett
bud = p_1 * x_1 + p_2 * x_2

# Beregne marginal substitusjonsrate (MSB)
msb = sp.diff(U, x_1) / sp.diff(U, x_2)

# Løs likning (14) for x_2
x_2_rel = sp.solve(msb - p_1 / p_2, x_2)[0]

# Sett inn i budsjettligningen og løs for x_1
budsj = bud.subs(x_2, x_2_rel)
x_1_eq = sp.solve(budsj - m, x_1)[0]

# Løsning for x_2 ved likevekt
x_2_eq = x_2_rel.subs(x_1, x_1_eq)
```

```python
# steg 1 har vi løst x_1_eq og x_2_eq

# steg 2
U_max=U.subs([(x_1,x_1_eq),(x_2,x_2_eq)])

# Vi kan bruke tall i et oppslag for å regne numeriske verdier
# her er et eksempel
# la oss lage noen tall i et oppslag
in_dict = {"a":0.035,"b":0.965,"m":1766,"p_1":3.02,"p_2":1}

# Vi bruker disse tallene til å finne nytten som oppnås samt etterspørsel

nytte=sp.sympify(U_max).evalf(subs={a:in_dict['a'], b:in_dict['b'], m:in_dict['m'], p_1:in_dict['p_1'], p_2:in_dict['p_2']})
x_1_num=sp.sympify(x_1_eq).evalf(subs={a:in_dict['a'], b:in_dict['b'], m:in_dict['m'], p_1:in_dict['p_1'], p_2:in_dict['p_2']})
x_2_num=sp.sympify(x_2_eq).evalf(subs={a:in_dict['a'], b:in_dict['b'], m:in_dict['m'], p_1:in_dict['p_1'], p_2:in_dict['p_2']})

# steg 3 
# vi har laget x_2_ind_sol som løsning
# og vi har laget en funksjon ved lambdify: indiff_x_2(U_0,a,b,x_1)

# Ta a=.4, b=.25, m=2000, p_1=25, p_2=40

# finn verdien for U_max, x_1_eq og x_2_eq


nytte = U_max.subs([(a,0.049),(b,0.951),(m,1766),(p_1,3.02),(p_2,1)])
x_1_num = x_1_eq.subs([(a,0.049),(b,0.951),(m,1766),(p_1,3.02),(p_2,1)])
x_2_num = x_2_eq.subs([(a,0.049),(b,0.951),(m,1766),(p_1,3.02),(p_2,1)])

nytte2 = U_max.subs([(a,0.049),(b,0.951),(m,1766),(p_1,3.61),(p_2,1)])
x_1_num2 = x_1_eq.subs([(a,0.049),(b,0.951),(m,1766),(p_1,3.61),(p_2,1)])
x_2_num2 = x_2_eq.subs([(a,0.049),(b,0.951),(m,1766),(p_1,3.61),(p_2,1)])

display(nytte)
display(x_1_num)
display(x_2_num)

display(nytte2)
display(x_1_num2)
display(x_2_num2)
```

```python
x = np.linspace(0, 1000, 1000)

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_ylabel('$x_2$', loc='top', rotation=0)
ax.set_xlabel('$x_1$', loc='right')
ax.set_xlim(0, 1000)
ax.set_ylim(0, 2000)
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Før bompenger
indiff_before = np.where(x > 0, indiff_x_2(nytte.subs(x_1, x), 0.049, 0.951, x), 0)
ax.plot(x, indiff_before, label='$U_{max}$', color='blue')
ax.plot(x, budsjett(1766, 3.02, 1, x), label='Konsumentens budsjett før bompenger', color='orange')

# Etter bompenger
indiff_after = np.where(x > 0, indiff_x_2(nytte2.subs(x_1, x), 0.049, 0.951, x), 0)
ax.plot(x, indiff_after, label='$U_{max}$', color='green')
ax.plot(x, budsjett(1766, 3.61, 1, x), label='Konsumentens budsjett etter bompenger', color='red')

# Tilpasningen
ax.vlines(float(x_1_num), 0, float(x_2_num), color='black', ls='dashed', label='$x_1$')
ax.hlines(float(x_2_num), 0, float(x_1_num), color='black', ls='dashed', label='$x_2$')

ax.vlines(float(x_1_num2), 0, float(x_2_num2), color='black', ls='dashed', label='$x_1$')
ax.hlines(float(x_2_num2), 0, float(x_1_num2), color='black', ls='dashed', label='$x_2$')

# Rutete svart bakgrunn
ax.grid(color='lightgrey', linestyle=':', linewidth=0.5)

# Tittel
ax.set_title('Konsumentens tilpasning')

# Vis navnene
ax.legend(bbox_to_anchor=(1.1, 0.35))

plt.ylim(1600, 1750)
plt.xlim(15, 40)


plt.gca().set_facecolor((0, 0, 0, 0.1))  # Gjennomsiktig svart bakgrunn
plt.show()
```

Cobb-Douglas-modellen
For å utvikle en Cobb-Douglas-modell trenger vi informasjon om kostnader, budsjett og preferanser. Vi har allerede bestemt estimater for kostnader og budsjett gjennom budsjettbetingelsen. Imidlertid er det ikke mulig å estimere preferansene direkte, derfor har vi valgt verdier som gir et realistisk bilde av virkeligheten i oppgaven. For å oppnå dette har vi brukt verdier for preferanser som er i samsvar med kjente opplysninger om bilister i Tromsø. Statistisk sentralbyrå (SSB) rapporterer at gjennomsnittlig kjørelengde for en bilist i Tromsø er 10 479 km per år, som tilsvarer 28,7 km per dag. Ved å bruke en preferanseverdi på 0,049 for gode x1 og 0,951 for gode x2, finner vi at konsumenten tilpasser seg tilnærmet nøyaktig til en daglig kjørelengde på 28,7 km.

Etterspørselen påvirkes av flere faktorer, hvorav pris er en av de mest betydningsfulle. Vanligvis vil etterspørselen synke når prisen øker. Dette blir tydelig illustrert i figuren. Før innføringen av bompenger kjørte konsumenten 28,7 kilometer og brukte 1679,5 kroner på andre varer. Etter innføringen av bompenger tilpasser konsumenten seg ved å kjøre 23,4 kilometer, samtidig som forbruket av andre varer forblir uendret. Dette indikerer en reduksjon på omtrent 18,5% i kjørelengden som etterspørres av konsumenten. Konsumenten befinner seg nå på en ny indifferenskurve med en lavere nyttenivå.

Bemerkelsesverdig er det at endringer i prisen på gode 1 (pris per kilometer) ikke påvirker forbruket av gode 2 (andre varer). Dette tyder på en uavhengighet mellom godene når det gjelder forbruk. Denne uavhengigheten er karakteristisk for Cobb-Douglas-funksjoner, da det ikke er noen krysspriselastisitet, men det kan også betraktes som en begrensning ved modellen.

Elastisiteten i etterspørselen etter kjørelengde avhenger spesielt av substitusjonsmulighetene i markedet, det vil si i hvilken grad personbiltransport kan erstattes med andre transportmidler, for eksempel kollektivtransport. Elastisiteten avhenger også av hvor stor andel av budsjettet som brukes til å konsumere godet. Jo større andel av budsjettet som brukes, desto mer elastisk blir etterspørselen.

Ettersom etterspørselen etter kjørelengde synker, øker naturligvis etterspørselen etter alternative varer (forutsatt alt annet er likt). Dette skyldes at de alternative varene nå blir relativt billigere, og fenomenet kalles substitusjonseffekten. Økningen i pris påvirker også konsumentens kjøpekraft, noe som reflekteres i inntektseffekten. 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1000, 1000)

fig, ax = plt.subplots()
ax.set_ylabel('$x_2$', fontsize=12)
ax.set_xlabel('$x_1$', fontsize=12)
ax.set_xlim(0, 1000)
ax.set_ylim(0, 2000)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#Før bompenger
ax.plot(x, indiff_x_2(nytte, 0.049, 0.951, x), color='blue', label='Indifferenskurve før bompenger')
ax.plot(x, budsjett(1766, 3.02, 1, x), color='green', label='Budsjettkurve før bompenger')

#Etter bompenger
ax.plot(x, indiff_x_2(nytte2, 0.049, 0.951, x), color='orange', label='Indifferenskurve etter bompenger')
ax.plot(x, budsjett(1766, 3.61, 1, x), color='red', label='Budsjettkurve etter bompenger')

#Forskyvning av budsjett
komp = nytte / nytte2
nyttbudsjett = 1766 * komp

ax.plot(x, budsjett(nyttbudsjett, 3.61, 1, x), color='purple', linestyle='dotted', label='Kompensert budsjettkurve')

x_1_num_komp = x_1_eq.subs([(a, 0.049), (b, 0.951), (m, nyttbudsjett), (p_1, 3.61), (p_2, 1)])
x_2_num_komp = x_2_eq.subs([(a, 0.049), (b, 0.951), (m, nyttbudsjett), (p_1, 3.61), (p_2, 1)])

#Tilpasningen
ax.vlines(x_1_num, 0, x_2_num, color='black', ls='dashed', label='$x_1$ før bompenger')
ax.hlines(x_2_num, 0, x_1_num, color='black', ls='dashed', label='$x_2$ før og etter bompenger')

ax.vlines(x_1_num2, 0, x_2_num2, color='red', ls='dashed', label='$x_1$ etter bompenger')
ax.hlines(x_2_num2, 0, x_1_num2, color='black', ls='dashed')

ax.vlines(x_1_num_komp, 0, x_2_num_komp, color='blue', ls='dashed', label='$x_1$ kompensert')
ax.hlines(x_2_num_komp, 0, x_1_num_komp, color='blue', ls='dashed', label='$x_2$ kompensert')

#Tittel
ax.set_title('Effekter av substitusjon og inntekt', fontsize=14)

#Vis navnene
ax.legend(bbox_to_anchor=(1.1, 0.35))

#Punkter på graf
ax.annotate('A', xy=(x_1_num, x_2_num + 10), size=10)
ax.annotate('B', xy=(x_1_num2 - 0.05, x_2_num2 + 10), size=10)
ax.annotate('C', xy=(x_1_num_komp, x_2_num_komp + 10), size=10)

plt.ylim(1650, 1710)
plt.xlim(22, 30)


plt.show()

display(x_1_num_komp)
display(x_2_num_komp)
```

I grafen vi har presentert, visualiserer vi forskjellige kurver og punkter som representerer effekter av substitusjon og inntekt. Ved å analysere grafen kan vi observere og forstå hvordan forbrukerens valg og preferanser endres i ulike scenarier.

Vi starter med å identifisere punkt A og punkt B på grafen. Bevegelsen fra punkt A til B representerer en endring i situasjonen eller betingelsene. I denne bevegelsen observerer vi substitusjonseffekten.

Substitusjonseffekten viser seg gjennom endringen i indifferenskurvene før og etter innføringen av bompenger. Før bompenger har vi en indifferenskurve og en budsjettkurve. Etter innføringen av bompenger ser vi en ny indifferenskurve og en justert budsjettkurve.

Ved å sammenligne kurvene og punktene før og etter bompenger, kan vi observere hvordan forbrukeren foretar substitusjoner mellom gode 1 og gode 2. Substitusjonseffekten viser seg ved at forbrukeren endrer sitt konsummønster for å opprettholde sin ønskede nytte eller tilfredshet. Dette kan sees gjennom justeringene i indifferenskurvene og de tilhørende punktene på grafen.

Det er verdt å merke seg at substitusjonseffekten ikke er den eneste faktoren som påvirker forbrukerens valg. Det er også andre økonomiske faktorer som spiller inn, for eksempel inntektsendringer og tilgjengeligheten av alternative varer eller tjenester. Disse faktorene kan også bidra til endringer i forbrukerens preferanser og valg.

Grafen gir oss en visuell representasjon av substitusjonseffekten og tillater oss å observere og analysere hvordan forbrukerens atferd endrer seg i respons til endringer i økonomiske betingelser. Den gir oss innsikt i dynamikken i markedet og forbrukerens reaksjoner på ulike økonomiske påvirkninger.

```python
substitusjonseffekt = x_1_num - x_1_num2
inntektseffekt = x_1_num2 - x_1_num_komp
totaleffekt = x_1_num - x_1_num_komp

print("Substitusjonseffekten: ", substitusjonseffekt)
print("Inntektseffekten: ", inntektseffekt)
print("Totaleffekten: ", totaleffekt)
```

Denne delen av koden beregner og viser tre ulike effekter: substitusjonseffekt, inntektseffekt og totaleffekt.

Substitusjonseffekten beregnes ved å ta differansen mellom verdien av x1 etter bompenger (x_1_num2) og verdien av x1 før bompenger (x_1_num). Denne effekten måler endringen i forbrukerens valg av x1 som en respons på prisendringen.

Inntektseffekten beregnes ved å ta differansen mellom verdien av x1 etter bompenger (x_1_num2) og verdien av x1 etter kompensasjon (x_1_num_komp). Denne effekten måler endringen i forbrukerens valg av x1 som en respons på endringen i inntekt.

Totaleffekten beregnes ved å ta differansen mellom verdien av x1 før bompenger (x_1_num) og verdien av x1 etter kompensasjon (x_1_num_komp). Denne effekten måler den samlede endringen i forbrukerens valg av x1 som følge av både prisendringen og endringen i inntekt.

Disse effektene blir deretter skrevet ut, slik at du kan se verdien av hver effekt.
