```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
def C_X(X):
    return (500*0.8367*X+350*1.1952*X+10000)/X
```

```python
x_num=np.linspace(0.001,100, 100)

fig2, ax = plt.subplots()
ax.set_ylabel('NOK', loc='top')
ax.set_xlabel('Produserte enheter', loc='right')
ax.set(xlim=(0,100))
ax.set(ylim=(0,3000))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
                

# plott funksjonen her

ax.plot(x_num, C_X(x_num), label='AC på lang sikt')

# tittel
ax.set_title('AC på kort og lang sikt')

#vis navnene:
ax.legend(loc='best',frameon=False);
```

```python
# skriv ditt svar her
def C_s(X,K):
    return (10000+500*K+(350*X**2)/K)/X



ax.plot(x_num, C_s(x_num, 10), label = '(K=10)', color='orange')
ax.plot(x_num, C_s(x_num, 25), label = ' (K=25)', color = 'darkred')
ax.plot(x_num, C_s(x_num, 50), label = '(K=50)', color= 'purple')
ax.legend(loc='best',frameon=False)
fig2
```
