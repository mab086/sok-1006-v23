```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
x, N, K , w , r = sp.symbols('x, N K w r ', positive=True, real=True)

def prod(N,K):
    produktfunksjon= N**(0.5)*K**(0.5)
    return produktfunksjon

prod(N,K)

def kost(N,K,w,r):
    kostnadsfunksjon= N*w + K*r + 10000
    return kostnadsfunksjon

kost(N,K,w,r)

n_sol = sp.solve(prod(N,K)- x, N)[0]
n_sol
```

```python
kost(n_sol,K,350,500)
```

```python
def fn(K,x):
    return 500*K+10000+(350*x**2)/K

fn(K,x)
```

```python
x_num = np.linspace(0.1,100,100)

fig1, ax = plt.subplots()
ax.spines['top']
ax.spines['right']

ax.plot(x_num, fn(10,x_num), label = 'Kostnandsfunksjon (K=10)', color='purple')
ax.plot(x_num, fn(50,x_num), label = "Kostnandsfunksjon (K=50)", color='darkred')

ax.legend(loc = 'best', frameon = False);
```
