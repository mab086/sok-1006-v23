```python
# importer pakkene som vi trenger
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
# definer symboler
N, A = sp.symbols('N A', positive=True, real=True)
```

```python


def prod(c,N,A):
    produksjon=1000*(c.exp(N)/(A+c.exp(N)))
    return produksjon
prod(sp,N,A)
```

```python
mp = sp.simplify(sp.diff(prod(sp,N,200),N))
display(mp)

mp2 = sp.lambdify((N), mp)
```

```python
def ap(c,N,A):
    gjennomsnitt=prod(c,N,A)/N
    return gjennomsnitt
ap(sp,N,A)
```

```python
n= np.linspace(1,10,100)

fig1, ax = plt.subplots()
ax.set_ylabel('Kvantum (x)', loc='top')
ax.set_xlabel('Enheter arbeidskraft (N)', loc='right')
ax.set(xlim=(1,10))
ax.set(ylim=(0,300))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

ax.plot(n,mp2(n), label = "Marginalproduktivitet", color='purple')
ax.plot(n,ap(np,n,200), label = "Gjennomsnittsproduktivitet", color='darkred')
ax.legend(loc = 'best', frameon = False);
ax.grid()
ax.legend(bbox_to_anchor = (1, 0.3))
```
